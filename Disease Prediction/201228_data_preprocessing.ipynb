```
disease_list = ['C16','C18','C34','H25-26','H40-42','K20-31','K58','K70']
remove_list =[]
before_id = id = df['YEAR'][0]+df['PERSON_ID'][0]
for i in range(len(df)-1):
  id = df['YEAR'][i+1]+df['PERSON_ID'][i+1]
  if id == before_id:
    remove_list.append(i)
    for d in disease_list:
      df[d][i+1] += df[d][i]
  else:
    before_id = id
df_drop = df.drop(df.index[remove_list])
```

```
norm_target = ['SEX', 'AGE', 'HEIGHT', 'WEIGHT', 'BP_HIGH',
       'BP_LWST', 'BLDS', 'TOT_CHOLE', 'HMG', 'OLIG_PROTE_CD', 'SGOT_AST',
       'SGPT_ALT', 'GAMMA_GTP', 'SMK_STAT_TYPE_RSPS_CD', 'SMK_TERM_RSPS_CD',
       'DSQTY_RSPS_CD', 'DRNK_HABIT_RSPS_CD', 'TM1_DRKQTY_RSPS_CD']

for t in norm_target:
  t_max = max(df[t])
  t_min = min(df[t])
  for i in range(len(df)):
    df[t][i] = (df[t][i]-t_min)/(t_max-t_min)
```

```
df_array = df[['SEX', 'AGE', 'HEIGHT', 'WEIGHT', 'BP_HIGH',
       'BP_LWST', 'BLDS', 'TOT_CHOLE', 'HMG', 'OLIG_PROTE_CD', 'SGOT_AST',
       'SGPT_ALT', 'GAMMA_GTP', 'FMLY_HPRTS_PATIEN_YN', 'FMLY_APOP_PATIEN_YN',
       'FMLY_HDISE_PATIEN_YN', 'FMLY_DIABML_PATIEN_YN',
       'FMLY_CANCER_PATIEN_YN', 'SMK_STAT_TYPE_RSPS_CD', 'SMK_TERM_RSPS_CD',
       'DSQTY_RSPS_CD', 'DRNK_HABIT_RSPS_CD', 'TM1_DRKQTY_RSPS_CD',
       'HCHK_APOP_PMH_YN', 'HCHK_HDISE_PMH_YN', 'HCHK_HPRTS_PMH_YN',
       'HCHK_DIABML_PMH_YN', 'HCHK_HPLPDM_PMH_YN', 'HCHK_ETCDSE_PMH_YN',
       'HCHK_PHSS_PMH_YN', 'DFAB_1', 'DFAB_2', 'DFAB_3', 'DFAB_4', 'DFAB_5',
       'DFAB_6', 'DFAB_7', 'DFAB_9', 'C16', 'C18', 'C34', 'H25-26', 'H40-42',
       'K20-31', 'K58', 'K70']]
```

```
df_array = df_array.to_numpy()
```

```
df_array
```

```
disease = 43 # K20-31
y = df_array[:,disease]
X = np.append(df_array[:,:disease],df_array[:,disease+1:],axis=1)

print(X)
print(y)
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state=1234)
```

```
from tensorflow.keras.layers import Dense, GRU, Embedding
from tensorflow.keras.models import Sequential
```

```
model = Sequential()
model.add(Embedding(45, 100))
model.add(GRU(128))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['acc'])
history = model.fit(X_train, y_train, epochs=4, validation_split=0.2)
```

```
print("\n 테스트 정확도: %.4f" % (model.evaluate(X_test, y_test)[1]))
```
